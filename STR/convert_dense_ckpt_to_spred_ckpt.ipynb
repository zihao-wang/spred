```python
import torch

import os
```

```python
dense_file = "pretrain/{}-Dense.pth"
spred_file = "pretrain/{}-spred-template.pth"
output_file = "pretrain/{}-spred.pth"

arch_list = ["ResNet50", "MobileNetV1"]
arch = "ResNet50"
```

```python
source_raw = torch.load(dense_file.format(arch))
source = source_raw['state_dict']
target_raw = torch.load(spred_file.format(arch))
if 'state_dict' in target_raw:
    target = target_raw['state_dict']
else:
    target = target_raw
```

```python
state_dict = {}
for long_name in source:
    short_name = long_name.replace("module.", "")    
    state_dict[long_name] = source[long_name]
    print("identify the name", long_name)

    short_red_name = short_name.replace(".weight", ".red_weight")
    if '.weight' in short_name and short_red_name in target:
        sign = torch.sign(source[long_name])
        state_dict[long_name] = sign * source[long_name].abs() ** 0.5 
        red_weight = source[long_name].abs() ** 0.5 
        long_red_name = long_name.replace(".weight", ".red_weight")
        state_dict[long_red_name] = red_weight
        #print(k)
        print("identify the red weight", long_red_name)

source_raw['state_dict'] = state_dict
print(output_file.format(arch))
torch.save(source_raw, output_file.format(arch))
```

```python
for k in source:
    sort_source = source[k].abs().flatten().sort()[0]
    i = int(len(sort_source) * 0.95)
    print(k, sort_source[i])
```
